# GPT2

```python
!python /content/drive/MyDrive/nlp-knowledge-graphs/extract.py /content/drive/MyDrive/nlp-knowledge-graphs/examples/candidates_movies.txt /content/drive/MyDrive/nlp-knowledge-graphs/gpt2-large-candidates_movies-1.jsonl --language_model gpt2-large --use_cuda true
```

```python
gpt2_path = project + 'gpt2-large-candidates_movies-1.jsonl'
source_path = project + 'examples/candidates_movies.txt'

with open(source_path, 'r') as f:
  source = f.read().split('\n')
  
gpt2_jsonObj = pd.read_json(gpt2_path, lines=True)
```

```python
gpt2_jsonObj['source'] = source[0:len(gpt2_jsonObj)]
```

```python
gpt2_output = project+'gpt2-large-candidates_movies-1.csv'
gpt2_jsonObj.to_csv(gpt2_output, index=False) 
```
